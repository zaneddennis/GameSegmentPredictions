```python
import pandas as pd
```

# Data Import

```python
df = pd.read_csv("PBP/1016OklahomaCityGoldenState.csv", index_col=0)
df
```

# Data Exploration

```python
subs = df.loc[df.Home.str.contains("enters the game", na=False) | df.Away.str.contains("enters the game", na=False)]
subs
```

# Identify Substitutions

```python
toDrop = []
for i, j in enumerate(subs.iterrows()):
    ix, r = j
    
    if i != 0:
        time = r["Time"]
        prevTime = subs.iloc[i-1]["Time"]
        
        if time == prevTime:
            toDrop.append(ix)

toDrop
```

```python
subMoments = subs.drop(toDrop).index
subMoments
```
