```python
import pandas as pd
```

```python
abbreviations = pd.read_csv("abbreviations.csv", header=None, index_col=0)
abbreviations = abbreviations[1]
abbreviations
```

```python
games = pd.read_csv("games_2018_2019.csv")
games = games.iloc[:1230]  # remove playoffs
games
```

```python
def GetGamePBP(row):
    m = str(row["Date"])[:-2]
    if len(m) == 1:
        m = "0" + m
    
    d = str(row["Date"])[-2:] + "0"
    if len(d) == 2:
        d = "0" + d
    
    y = "2018" if int(m) >= 10 else "2019"
    t = abbreviations[row["H"]]
    
    url = "https://www.basketball-reference.com/boxscores/pbp/{}{}{}{}.html".format(y, m, d, t)
    print(url)
    
    pbp = pd.read_html(url)[0]
    pbp.columns = ["Time", "Away", "A_Delta", "Score", "H_Delta", "Home"]
    pbp = pbp.iloc[1:].reset_index(drop=True)
    return pbp[["Time", "Away", "Score", "Home"]]

GetGamePBP(games.iloc[0])
```

```python
for i, r in games.iterrows():
    df = GetGamePBP(r)
    df.to_csv("PBP/" + str(r["Date"]) + r["V"] + r["H"] + ".csv")
```
